## ***IMPORTING LIBRARIES***

```python
import numpy as np
import pandas as pd
```

## ***LOADING THE DATASET***

```python
data = pd.read_csv("Breast Cancer Prediction.csv")
```

## ***SUMMARIZING THE DATASET***

```python
print(data.shape)
print(data.head())
```

## ***SEGREGATING THE DATASET INTO INPUT AND OUTPUT***

```python
input  = data.iloc[:,:-1].values
print(input)
output  = data.iloc[:,-1].values
print(output)
```

## ***SPLITTING THE DATASET INTO TRAIN AND TEST DATA***

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(input,output)
```

## ***TRAINING THE DATASET***

```python
from xgboost import XGBClassifier 

XGB_Model = XGBClassifier()

XGB_Model.fit(x_train,y_train)
```

## ***CALCULATING THE ACCURACY OF THE TRAINED MODEL***

```python
y_pred = XGB_Model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

print("accuracy_score :", accuracy_score(y_test,y_pred)*100,"%")
print("confusion_matrix :")
confusion_matrix(y_test,y_pred)
```
